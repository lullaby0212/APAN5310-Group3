# Create region_id, ship it off to the Region table, and then add region_id to the dataset

```python
unique_regions = df['Region'].unique()
temp_region_df = pd.DataFrame({'name': unique_regions})
temp_region_df.insert(0, 'region_id', range(1, 1 + len(temp_region_df)))

display(temp_region_df)
```

```python
temp_region_df.to_sql(name='REGION', con=engine, if_exists='append', index=False)
```

```python
region_id_list = [temp_region_df.region_id[temp_region_df.name == i].values[0] for i in df.Region]

df.insert(8, 'region_id', region_id_list)

df
```

# Create country id, ship it off into the Country table, and then add the country_id key to the dataset

```python
# Create temp_country_df with unique countries and their IDs
temp_country_df_base = pd.DataFrame({'name': df.Country.unique()})
temp_country_df_base.insert(0, 'country_id', range(1, 1 + len(temp_country_df_base)))

# Get unique Country and region_id pairings from df
unique_country_region_id = df[['Country', 'region_id']].drop_duplicates()

# Merge temp_country_df_base with unique_country_region_id to add the region_id column
temp_country_df = pd.merge(
    temp_country_df_base,
    unique_country_region_id,
    left_on='name',
    right_on='Country',
    how='left'
)

# Drop the redundant 'Country' column after merge
temp_country_df = temp_country_df.drop(columns=['Country'])

display(temp_country_df)
```

```python
temp_country_df.to_sql(name='COUNTRY', con=engine, if_exists='append', index=False)
```

```python
country_id_list = [temp_country_df.country_id[temp_country_df.name == i].values[0] for i in df.Country]

df.insert(8, 'country_id', country_id_list)

df.head()
```

# Create team_id, ship it off to the Team table, and then add it to the dataset

```python
unique_team = df['Team'].unique()
sorted_unique_team = sorted(unique_team)
temp_team_df = pd.DataFrame({'name': sorted_unique_team})
temp_team_df.insert(0, 'team_id', range(1, 1 + len(temp_team_df)))

temp_team_df
```

```python
temp_team_df.to_sql(name='TEAM', con=engine, if_exists='append', index=False)
```

```python
team_id_list = [temp_team_df.team_id[temp_team_df.name == i].values[0] for i in df.Team]

df.insert(3, 'team_id', team_id_list)

df
```

# Create employee_id, compile everything to be shipped into Employee table, and then add employee_id to the dataset

```python
# Step 1: Create a mapping of unique Sales_Rep to a new employee_id
unique_sales_reps = df['Sales_Rep'].unique()
employee_mapping_df = pd.DataFrame({'full_name': unique_sales_reps})
employee_mapping_df.insert(0, 'employee_id', range(1, 1 + len(employee_mapping_df)))

# Step 2: Extract all unique (Sales_Rep, team_id, region_id) pairings from df
# Using drop_duplicates() without a subset ensures that if a Sales_Rep has
# worked in multiple team_id/region_id combinations, all those unique combinations are kept.
all_employee_pairings = df[['Sales_Rep', 'team_id']].drop_duplicates()

# Step 3: Merge to create the final temp_employee_df with employee_id and all pairings
# First, rename 'Sales_Rep' in all_employee_pairings to 'full_name' for merging
all_employee_pairings = all_employee_pairings.rename(columns={'Sales_Rep': 'full_name'})

temp_employee_df = pd.merge(
    employee_mapping_df,
    all_employee_pairings,
    on='full_name',
    how='left'
)

# Display the result
display(temp_employee_df)
```

```python
temp_employee_df.to_sql(name='EMPLOYEE', con=engine, if_exists='append', index=False)
```

```python
copy_employee_mapping_df = employee_mapping_df.rename(columns={'full_name': 'Sales_Rep'})
copy_employee_mapping_df
```

```python
# Merge df with employee_mapping_df to add the employee_id
df = pd.merge(
    df,
    copy_employee_mapping_df[['employee_id', 'Sales_Rep']],
    on='Sales_Rep',
    how='left'
)

df
```

# Product Category

```python
df_prod_cat = (
    df[["Product_Category"]]
    .drop_duplicates()
    .reset_index(drop=True)
    .rename(columns={"Product_Category": "category_name"})
)

df_prod_cat["category_id"] = range(1, len(df_prod_cat) + 1)
df_prod_cat = df_prod_cat[["category_id", "category_name"]]
```

```python
df = pd.merge(
    df,
    copy_df_prod_cat[['category_id', 'Product_Category']],
    on='Product_Category',
    how='left'
)

df
```

# Product

```python
df_product = pd.DataFrame({'product_name': df.Product_Name.unique()})
df_product.insert(0, 'product_id', range(1, 1 + len(df_product)))

product_category_pairing = df[['Product_Name', 'category_id']].drop_duplicates()

# Step 3: Merge to create the final temp_employee_df with employee_id and all pairings
# First, rename 'Sales_Rep' in all_employee_pairings to 'full_name' for merging
product_category_pairing = product_category_pairing.rename(columns={'Product_Name': 'product_name'})

temp_product_df = pd.merge(
    df_product,
    product_category_pairing,
    on='product_name',
    how='left'
)

temp_product_df
```

```python
temp_product_df.to_sql(name='PRODUCT', con=engine, if_exists='append', index=False)
```

```python
copy_df_product_for_db = temp_product_df.rename(columns={'product_name': 'Product_Name'})
copy_df_product_for_db
```

```python
df = pd.merge(
    df,
    copy_df_product_for_db[['product_id', 'Product_Name']],
    on='Product_Name',
    how='left'
)

df
```
